## Problem 1a

The formula for normal distribution is shown below:

Probability-density-function-of-Normal-Distribution.jpg.webp

```python
import numpy as np
import matplotlib.pyplot as plt

class normal_distribution():
  def __init__(self, x, mu, std):
    #listing parameters to initiate normal distribution
    self.x = x
    self.mu = mu
    self.std = std
    #self.n_size = n_size
  
  def pdf(self):
    if self.std <= 0:
      raise Exception('sigma cannot be less than or equal to zero')
    prob_density = 1/np.sqrt(2 * np.pi * self.std**2) * np.exp(-(self.x-self.mu)**2/ (2*(self.std**2)))
    return prob_density

  def plotpdf(self):
    plt.scatter(self.x, normal_distribution.pdf(self))
    plt.plot(self.x, normal_distribution.pdf(self))
    plt.title("Normal Distribution") 
    plt.xlabel("Random variable")
    plt.ylabel("Probability density")
    plt.show()


#calling the function:
#Parameters for data:
mu = 7
std = 2
  #For normal dist: 99.9% data is within 4 standard deviation. Hence, range is set to +/- 4* sigma
min_range = mu - 4*std
max_range = mu + 4*std

#from problem statement:
n_size1 = 100
n_size2 = 1000
n_size3 = 100000

#Generate data for x axis:
x1 = np.linspace(min_range, max_range, n_size1) 
x2 = np.linspace(min_range, max_range, n_size2) 
x3 = np.linspace(min_range, max_range, n_size3) 

#calling function
pdf1 = normal_distribution(x1, mu, std)
pdf2 = normal_distribution(x2, mu, std)
pdf3 = normal_distribution(x3, mu, std)

#show plot
plot1 = pdf1.plotpdf()
plot2 = pdf2.plotpdf()
plot3 = pdf3.plotpdf()
```

## Problem 1b
The three normal distribution probability density function shapes are exactly the same for all of the three sample size as the mean and standard deviation are the same. Scatter plot is used to show the actual data points. The only difference are the curve are smoother (closer to 1 smooth line) as the sample size increases. 

Only when the sample size is low (e.g. n = 10) then we can see non-smooth curve as shown below (for reference only).

```python
x0 = np.linspace(min_range, max_range, 10) 
pdf0 = normal_distribution(x0, mu, std)
plot0 = pdf0.plotpdf()
```
